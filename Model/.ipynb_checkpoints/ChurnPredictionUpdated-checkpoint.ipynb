```python
import seaborn as sn
import pandas as pd
import numpy as np
import warnings
import joblib


from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, cross_val_score, GridSearchCV, train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
warnings.filterwarnings('ignore')
```

```python
# Creating a dataframe

df = pd.read_csv("Churn_Modelling.csv")
df
```

```python
# Dropping the non-required columns

df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 'columns', inplace = True) 
```

```python
df.dtypes
```

```python
# Checking if any null values are present

nan_values = df[df.isnull().any(axis=1)]  
nan_values
```

```python
df
```

```python
# Visualization of the exited and not-exited customers with respect to the tenure

not_exited_tenure = df[df.Exited == 0].Tenure
exited_tenure = df[df.Exited == 1].Tenure

plt.xlabel("Tenure")
plt.ylabel("No. of customers")
plt.title("Customer Churn Prediction Visualization - Tenure")

plt.hist([not_exited_tenure, exited_tenure], color = ['green', 'red'], label = ['Not Exited', 'Exited'])
plt.legend()
```

```python
# Visualization of the exited and not-exited customers with respect to the expected Salary

not_exited_salary = df[df.Exited == 0].EstimatedSalary
exited_salary = df[df.Exited == 1].EstimatedSalary

plt.xlabel("Salary")
plt.ylabel("No. of customers")
plt.title("Customer Churn Prediction Visualization - Salary")

plt.hist([not_exited_salary, exited_salary], color = ['green', 'red'], label = ['Not Exited', 'Exited'])
plt.legend()
```

```python
# Checking for all the unique values in each column for encoding

for column in df:
        print(f'{column} : {df[column].unique()}')
```

```python
# Replacing the Male and Female values with 0 and 1 respectively

df.Gender.replace(['Female', 'Male'], [1, 0], inplace = True)
df.Gender.unique()
```

```python
# Converting the categorical values in the Geography column to the numerical values using the One Hot Encoding with pandas

dummies = pd.get_dummies(df.Geography, columns = ['Geography'], drop_first = True)
dummies = dummies.astype(int)
df.drop('Geography', axis = 'columns', inplace = True)
df = pd.concat([df, dummies], axis = 1)
df
```

```python
df.dtypes
```

```python
# Scaling our dataset from 0 till 1 to improve the efficiency

columns_to_scale = ['CreditScore', 'Age', 'Tenure', 'NumOfProducts', 'Balance', 'EstimatedSalary']

scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
df
```

```python
# Checking the distribution of the exited and not-exited customers

df.Exited.value_counts()
```

```python
# Due to the biased or imbalanced distribution, using the SMOTE method to sample the data for the better precision and recall

X = df.drop('Exited', axis = 1)
y = df.Exited

from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy = 'minority')
X_smote, y_smote = smote.fit_resample(X, y)

X_smote.value_counts(), y_smote.value_counts()
```

```python
# Defining classifiers and their parameters to use in the GridSearchCV

random_forest = RandomForestClassifier()
random_forest_params = {
    'n_estimators': [50, 100],
    'criterion': ['gini', 'entropy']
    }
```

```python
logistic_regression = LogisticRegression()
logistic_regression_params = {
    'penalty': ['elasticnet', 'none'],
    'C': [0.1, 1.0, 10.0]
}
```

```python
gradient_boosting = GradientBoostingClassifier()
gradient_boosting_params = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1]
}
```

```python
classifiers = [
    (random_forest, random_forest_params),
    (logistic_regression, logistic_regression_params),
    (gradient_boosting, gradient_boosting_params)
]
```

```python
# Using the GridSearchCV to analyze the result with the different classifiers and their parameters

scores = []

for classifier, params in classifiers:
    grid_search = GridSearchCV(classifier, params, cv = 5)
    grid_search.fit(X_smote, y_smote)
    
    scores.append({
        'model' : classifier,
        'best_score' : grid_search.best_score_,
        'best_params' : grid_search.best_params_
    })
```

```python
# From the result of the GridSearchCV getting the classifier with the best accuracy with its parameters

df = pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])
df
```

```python
# Separating the training and testing data

X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size = 0.2, stratify = y_smote)
X_train.shape
```

```python
# Training the model with the best classifier that we got from the GridSearchCV 

random_forest = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
model = random_forest.fit(X_train, y_train)
```

```python
# Getting the accuracy of the model from the testing data

model.score(X_test, y_test)
```

```python
# To further analyze our model, storing the predicted data of the testing sample into a variable

y_predicted = model.predict(X_test)
y_predicted
```

```python
# Generating a classification report

print(classification_report(y_test, y_predicted))
```

```python
# Displaying a confustion matrix

confusion_matrix = metrics.confusion_matrix(y_test, y_predicted)

plt.figure(figsize = (10, 7))
sn.heatmap(confusion_matrix, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Storing our model in a joblib file

joblib.dump(model, "churn_prediction.joblib")
```

```python
# Importing the model from the joblib file and testing if it works

new_model = joblib.load("churn_prediction.joblib")
new_model.predict(X_test)
```
